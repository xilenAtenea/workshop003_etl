```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from db_queries import get_real_data_and_predic_from_db
```

```python
# Take real data and predictions from the DB 

df = get_real_data_and_predic_from_db()
print(df.head(5))
```

```python
# Setting variables to the data

y_test = df['happiness_score']
predictions = df['happiness_score_prediction']
```

```python
# Metrics

mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

rounded_mse = round(mse, 3)
rounded_mae = round(mae, 3)
rounded_r2 = (round(r2, 3))*100

print('mean_squared_error : ', rounded_mse)
print('mean_absolute_error : ', rounded_mae)
print(f'r_squared: {rounded_r2}%')
```

**mean_squared_error (MSE):** The value of MSE is 0.268. The MSE measures the mean squared differences between the model predictions and the actual values. A low MSE value indicates that the model is a good fit to the data because the differences between the predictions and the actuals are small on average. In this case, an MSE of 0.268 indicates that the model has a fairly good fit to the actual data.

**mean_absolute_error (MAE):** The MAE value is 0.409. The MAE measures the average of the absolute differences between the model predictions and the actual values. Similar to the MSE, a low MAE value indicates a good fit of the model to the data because the absolute differences are small on average. In this case, an MAE of 0.409 also indicates a good fit of the model to the actual data.

**r_squared:** The R-squared of 78.8% suggests that roughly 78.8% of the variability in the dependent variable is accounted for by the independent variables included in the model, i.e. the r_squared values reveals that 78.8% of the variability observed in the dependent variable is explained by the regression model.

```python
# Plotting actual model values vs. predictions

plt.figure(figsize = (20,10))

plt.plot(predictions, label='Prediction')
plt.plot(y_test.values, label='Real values')
plt.ylabel('count')
plt.legend()
plt.show()
```

The plot provides a visual comparison between the implemented model's predictions (line in blue) and the actual values (line in orange). Overall, the model demonstrates a close fit, closely aligning the prediction line with the actual values. This graphical representation serves as a tool to visually comprehend the differences between the predicted values generated by the model and the true observed values.
